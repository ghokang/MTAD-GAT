# Data Save

```python
import pandas as pd
import os
os.chdir("/Users/mac/Desktop/paper/code/MTAD-GAT/data/DATA")
```

```python
list_name = os.listdir()
os.chdir("/Users/mac/Desktop/paper/code")
```

```python
os.makedirs("./Total Result", exist_ok=True)
for i in list_name:
    if not i.endswith(".csv"):
        continue
    name = i.split(".")[0]
    result_path = "./MTAD-GAT/result/sub" + str(name) + "_iter_0_testresult.csv"
    if not os.path.exists(result_path):
        print(f"Skipping {name}: result file not found")
        continue
    predict = pd.read_csv(result_path)["Pred_Global"]
    data = pd.read_csv("./MTAD-GAT/data/DATA/" + str(name) + ".csv")
    result = pd.DataFrame({'Time':data.index[64:]+1})
    result = pd.concat([result, data.iloc[64:,:].reset_index(drop=True), predict],axis=1)
    result.to_csv("./Total Result/result_"+str(name)+".csv",index=False)
```
